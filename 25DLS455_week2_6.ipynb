**Introduction to Large Language Models (LLMs)**

```python
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer, BertForMaskedLM, BertTokenizer
import torch
import json

# Task 1: Text Generation using GPT-2
print("Task 1: Text Generation using GPT-2")
generator = pipeline("text-generation", model="gpt2")
prompt = "The future of AI is"
output = generator(prompt, max_new_tokens=30)[0]['generated_text']
print("Generated Text:\n", output)
print("\n" + "-"*80)

# Task 2: Experiment with Temperature
print("Task 2: Experiment with Temperature")
output_temp_02 = generator(prompt, max_new_tokens=30, temperature=0.2)[0]['generated_text']
output_temp_10 = generator(prompt, max_new_tokens=30, temperature=1.0)[0]['generated_text']
print("Temperature 0.2 Output:\n", output_temp_02)
print("\nTemperature 1.0 Output:\n", output_temp_10)
print("\n" + "-"*80)

# Task 3: Fill-in-the-Blank using BERT
print("Task 3: Fill-in-the-Blank using BERT")
fill_mask = pipeline("fill-mask", model="bert-base-uncased")
sentence = "The capital of India is [MASK]."
predictions = fill_mask(sentence)
print("Top 2 Predictions:")
for pred in predictions[:2]:
    print(f"{pred['sequence']} (score: {pred['score']:.4f})")
print("\n" + "-"*80)

# Task 4: Mock Tool Call Output
print("Task 4: Mock Tool Call Output")
tool_call = {
    "function": "get_weather",
    "arguments": {
        "city": "Delhi"
    }
}
print(json.dumps(tool_call, indent=2))
print("\n" + "-"*80)

# Task 5: Analyze Model Configuration
print("Task 5: Analyze Model Configuration")
model_name = "gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Count number of parameters
num_params = sum(p.numel() for p in model.parameters())
print(f"Number of parameters in {model_name}: {num_params}")

# Print model configuration
print("\nModel Configuration:")
print(model.config)
```
