P-1: Linear Algebra and Eigenvalues

```python
import numpy as np
import matplotlib.pyplot as plt

# Create matrices A (3x2) and B (2x3)
A = np.random.randint(1, 10, (3, 2))
B = np.random.randint(1, 10, (2, 3))

# Multiply A and B
C = np.dot(A, B)

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("A * B:\n", C)

# Define matrix D and compute eigenvalues and eigenvectors
D = np.array([[4, 2], [1, 3]])
eigenvalues, eigenvectors = np.linalg.eig(D)

print("Matrix D:\n", D)
print("Eigenvalues of D:", eigenvalues)
print("Eigenvectors of D:\n", eigenvectors)

# Plot eigenvectors using quiver
origin = np.zeros((2, 2))  # Two vectors starting at (0,0)
plt.figure(figsize=(6, 6))
plt.quiver(origin[0], origin[1], eigenvectors[0], eigenvectors[1],
           angles='xy', scale_units='xy', scale=1, color=['red', 'blue'])

plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.grid()
plt.title("Eigenvectors of Matrix D")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.gca().set_aspect('equal')
plt.show()
```

P-2: Symbolic Differentiation and Gradient Step

```python
import sympy as sp

# Define and differentiate f(x)
x = sp.symbols('x')
f = 3*x**3 - 5*x**2 + 2*x - 1
f_prime = sp.diff(f, x)

print("f(x):", f)
print("f'(x):", f_prime)

# Evaluate at x = 2
f_prime_val = f_prime.subs(x, 2)
print("f'(2) =", f_prime_val)

# Define f(x, y) = x^2 + y^2 and compute gradient
x, y = sp.symbols('x y')
f2 = x**2 + y**2
grad_f2 = [sp.diff(f2, var) for var in (x, y)]

print("Gradient of f(x, y) =", grad_f2)

# Gradient descent step from (x=3, y=4) with learning rate 0.1
x_val, y_val = 3, 4
learning_rate = 0.1
grad_values = [g.subs({x: x_val, y: y_val}) for g in grad_f2]

x_new = x_val - learning_rate * grad_values[0]
y_new = y_val - learning_rate * grad_values[1]

print("New (x, y) after one step:", (x_new, y_new))
```

P-3: Probability with Dice and Coin Simulation

```python
import numpy as np
import matplotlib.pyplot as plt

# Dice simulation
dice_rolls = np.random.randint(1, 7, size=1000)
plt.hist(dice_rolls, bins=np.arange(1, 8)-0.5, edgecolor='black', rwidth=0.8)
plt.xticks(range(1, 7))
plt.title("Histogram of Dice Rolls")
plt.xlabel("Dice Face")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

print("Empirical Mean:", np.mean(dice_rolls))
print("Empirical Variance:", np.var(dice_rolls))

# Biased coin toss
coin_tosses = np.random.choice(['Heads', 'Tails'], size=500, p=[0.7, 0.3])
unique, counts = np.unique(coin_tosses, return_counts=True)

# Plot bar chart
plt.bar(unique, counts, color=['blue', 'orange'])
plt.title("Biased Coin Toss")
plt.ylabel("Count")
plt.grid(True, axis='y')
plt.show()

print("Heads:", counts[unique.tolist().index('Heads')])
print("Tails:", counts[unique.tolist().index('Tails')])
```
